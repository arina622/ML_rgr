# Теоретическая часть

Нечёткая логика представляет собой расширение классической булевой алгебры, разработанное для формализации и моделирования рассуждений в условиях неопределённости и неточности исходных данных. В отличие от бинарной логики, оперирующей дихотомическими значениями «истина» или «ложь», нечёткая логика допускает частичную истинность высказываний, что позволяет отражать человеческие суждения и качественные описания сложных систем.

Основным объектом нечёткой логики является лингвистическая переменная. Под лингвистической переменной понимается переменная, значения которой определяются не числовыми величинами, а словами или предложениями естественного языка. Формально лингвистическая переменная характеризуется кортежем вида ⟨X, T, U, M⟩, где X – название переменной, T – множество её лингвистических значений (термов), U – универсальное множество (область определения), а M – семантическое правило, ставящее в соответствие каждому лингвистическому значению t из T нечёткое множество A на универсальном множестве U. В практических задачах в качестве лингвистических переменных выступают такие понятия, как «температура», «скорость», «надёжность», а их значениями – «высокий», «средний», «низкий».

Каждому лингвистическому значению лингвистической переменной ставится в соответствие функция принадлежности, которая количественно выражает степень соответствия элемента универсального множества U данному нечёткому множеству A. Функция принадлежности μ_A(x) отображает универсальное множество U в единичный интервал [0, 1], где значение 0 означает полное несоответствие элемента x нечёткому множеству A, значение 1 – полное соответствие, а промежуточные значения отражают различные степени частичной принадлежности. Наиболее распространёнными видами функций принадлежности являются треугольные, трапециевидные и гауссовы функции, выбор которых определяется спецификой решаемой задачи и требуемой точностью моделирования.

Процесс преобразования чётких входных значений в степени принадлежности к лингвистическим термам называется фаззификацией. На этапе фаззификации для каждого измеренного значения входной переменной вычисляются значения всех соответствующих функций принадлежности, что позволяет перейти от числового описания состояния системы к его лингвистической интерпретации.

Далее система нечёткого вывода оперирует базой нечётких правил, представляющей собой совокупность условных высказываний вида «ЕСЛИ (посылка), ТО (заключение)». Посылка и заключение формулируются с использованием лингвистических переменных. База правил формализует эмпирические знания эксперта о поведении моделируемой системы. Для каждого правила выполняется агрегация подусловий, в ходе которой с помощью операций нечёткой логики вычисляется степень истинности всего условия правила.

На этапе активации вычисленная степень истинности условия правила применяется к заключению. При этом функция принадлежности, соответствующая выходному лингвистическому терму в заключении правила, модифицируется. Наиболее часто применяются методы минимума, при котором функция принадлежности «обрезается» на уровне вычисленной степени истинности, или произведения, при котором функция принадлежности умножается на эту степень.

Результаты активации всех правил затем объединяются в единое нечёткое множество на этапе аккумуляция. Для объединения используются s-нормы, например, операция максимума. В итоге формируется результирующее нечёткое множество, описывающее выходную переменную системы в терминах нечёткой логики.

Для получения количественного, чёткого значения выходной переменной выполняется процедура дефаззификации. Дефаззификация представляет собой преобразование результирующего нечёткого множества в конкретное числовое значение. Существует несколько методов дефаззификации, выбор которых влияет на точность и динамические характеристики системы.

Наиболее распространённым методом является метод центра тяжести (центроида), при котором выходное значение вычисляется как абсцисса центра тяжести площади под кривой результирующей функции принадлежности. Этот метод обеспечивает плавное изменение выходного сигнала и учитывает вклад всех активированных правил. Формула для непрерывного универсального множества U имеет вид: y = (∫_U μ(y)* y dy)/(∫_U μ(y) dy). Для дискретного случая интегралы заменяются суммами.

Метод среднего максимума заключается в вычислении среднего арифметического значений, при которых результирующая функция принадлежности достигает своего максимума. Этот метод является менее вычислительно затратным, но может приводить к скачкообразным изменениям выходного сигнала при небольшом изменении входных данных, особенно если максимум функции принадлежности представляет собой плато.

Метод левого модального значения и метод правого модального значения выбирают, соответственно, наименьшее или наибольшее значение из множества, на котором достигается максимум функции принадлежности. Эти методы применяются в задачах, где критично минимальное или максимальное воздействие.

Выбор конкретного метода дефаззификации определяется требованиями к системе: необходимостью минимизации вычислений, важностью плавности управления, допустимостью дискретности выходного сигнала. В системах управления, как правило, предпочтение отдаётся методу центра тяжести вследствие его естественности и непрерывности результатов. Однако в реальных системах с ограниченными вычислительными ресурсами может быть выбран более простой метод.


# Практическая часть

## 1. Описание системы нечеткого вывода

### 1.1 Формализация входных и выходных переменных

Разрабатываемая система нечёткого вывода предназначена для оценки кредитоспособности заёмщика на основе трёх критериев. Входные переменные представлены в числовом виде, однако в рамках нечёткой модели они интерпретируются через лингвистические термы.

**Входная переменная 1**: Уровень дохода (income)

Универсальное множество: U_1 = [0, 200] тыс. рублей в месяц.

Лингвистические значения:
- Низкий: характеризует доходы, недостаточные для уверенного обслуживания кредитных обязательств;
- Средний: соответствует доходам, позволяющим обслуживать кредит при разумной долговой нагрузке;
- Высокий: указывает на доходы, обеспечивающие высокую платёжеспособность.

**Входная переменная 2**: Кредитная история (credit_history)

Универсальное множество: U_2 = [0, 100] условных баллов.

Лингвистические значения:
- Плохая: соответствует низкому баллу, указывает на наличие просрочек или дефолтов в прошлом;
- Средняя: отражает приемлемую историю с незначительными нарушениями;
- Отличная: характеризует безупречную историю исполнения обязательств.

**Входная переменная 3**: Возраст (age)

Универсальное множество: U_3 = [18, 90] лет.

Лингвистические значения:
- Молодой: возрастная группа с повышенным риском из-за недостатка кредитной истории и нестабильности доходов.
- Средний: наиболее предпочтительная группа с устоявшимся финансовым положением.
- Пожилой: группа с потенциальным снижением доходов в перспективе.

Выходная переменная: Рейтинг надёжности заёмщика (reliability)

Универсальное множество: U_4 = [0, 100] баллов.

Лингвистические значения:
- Очень низкий: указывает на неприемлемый уровень кредитного риска.
- Низкий: соответствует высокому риску, требующему дополнительных гарантий.
- Средний: характеризует приемлемый риск со стандартными условиями кредитования.
- Высокий: указывает на низкий уровень риска.
- Очень высокий: соответствует минимальному кредитному риску.

Для центральных лингвистических значений функции принадлежности для определены как треугольные (trimf) ввиду их вычислительной простоты и достаточной адекватности для задачи скоринга и трапециевидные (trapmf) для крайних лингвистических значений для корректной обработки граничных значения входных параметров. Параметры функций подобраны эмпирически на основе типичных представлений о границах лингвистических категорий в банковской практике.


Импорт необходимых библиотек

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Настройка отображения графиков

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['font.size'] = 12
```

Создание лингвистических переменных

```python
income_universe = np.arange(0, 201, 1)
history_universe = np.arange(0, 101, 1)
age_universe = np.arange(18, 91, 1)
reliability_universe = np.arange(0, 101, 1)
```

```python
income = ctrl.Antecedent(income_universe, 'доход')
credit_history = ctrl.Antecedent(history_universe, 'кредитная_история')
age = ctrl.Antecedent(age_universe, 'возраст')
reliability = ctrl.Consequent(reliability_universe, 'рейтинг_надёжности', defuzzify_method='centroid')
```

Функции принадлежности для переменной "Доход"

```python
income['низкий'] = fuzz.trapmf(income_universe, [0, 0, 40, 80])
income['средний'] = fuzz.trimf(income_universe, [60, 100, 140])
income['высокий'] = fuzz.trapmf(income_universe, [120, 160, 200, 200])

fig1, ax1 = plt.subplots()
ax1.plot(income_universe, income['низкий'].mf, 'b', linewidth=1.5, label='Низкий')
ax1.plot(income_universe, income['средний'].mf, 'g', linewidth=1.5, label='Средний')
ax1.plot(income_universe, income['высокий'].mf, 'r', linewidth=1.5, label='Высокий')
ax1.set_title('Функции принадлежности: Уровень дохода')
ax1.set_xlabel('Доход, тыс. руб./мес')
ax1.set_ylabel('Степень принадлежности')
ax1.legend()
ax1.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Функции принадлежности для переменной "Кредитная история"

```python
credit_history['плохая'] = fuzz.trapmf(history_universe, [0, 0, 30, 40])
credit_history['средняя'] = fuzz.trimf(history_universe, [30, 50, 70])
credit_history['отличная'] = fuzz.trapmf(history_universe, [60, 70, 100, 100])

fig2, ax2 = plt.subplots()
ax2.plot(history_universe, credit_history['плохая'].mf, 'b', linewidth=1.5, label='Плохая')
ax2.plot(history_universe, credit_history['средняя'].mf, 'g', linewidth=1.5, label='Средняя')
ax2.plot(history_universe, credit_history['отличная'].mf, 'r', linewidth=1.5, label='Отличная')
ax2.set_title('Функции принадлежности: Кредитная история')
ax2.set_xlabel('Баллы кредитной истории')
ax2.set_ylabel('Степень принадлежности')
ax2.legend()
ax2.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Функции принадлежности для переменной "Возраст"

```python
age['молодой'] = fuzz.trapmf(age_universe, [18, 18, 25, 35])
age['средний'] = fuzz.trimf(age_universe, [25, 45, 65])
age['пожилой'] = fuzz.trapmf(age_universe, [55, 65, 90, 90])

fig3, ax3 = plt.subplots()
ax3.plot(age_universe, age['молодой'].mf, 'b', linewidth=1.5, label='Молодой')
ax3.plot(age_universe, age['средний'].mf, 'g', linewidth=1.5, label='Средний')
ax3.plot(age_universe, age['пожилой'].mf, 'r', linewidth=1.5, label='Пожилой')
ax3.set_title('Функции принадлежности: Возраст заёмщика')
ax3.set_xlabel('Возраст, лет')
ax3.set_ylabel('Степень принадлежности')
ax3.legend()
ax3.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Функции принадлежности для переменной "Рейтинг надёжности"

```python
reliability['очень_низкий'] = fuzz.trimf(reliability_universe, [0, 10, 25])
reliability['низкий'] = fuzz.trimf(reliability_universe, [15, 30, 45])
reliability['средний'] = fuzz.trimf(reliability_universe, [35, 50, 65])
reliability['высокий'] = fuzz.trimf(reliability_universe, [55, 70, 85])
reliability['очень_высокий'] = fuzz.trimf(reliability_universe, [75, 90, 100])

fig4, ax4 = plt.subplots()
ax4.plot(reliability_universe, reliability['очень_низкий'].mf, 'r', linewidth=1.5, label='Очень низкий')
ax4.plot(reliability_universe, reliability['низкий'].mf, 'orange', linewidth=1.5, label='Низкий')
ax4.plot(reliability_universe, reliability['средний'].mf, 'y', linewidth=1.5, label='Средний')
ax4.plot(reliability_universe, reliability['высокий'].mf, 'g', linewidth=1.5, label='Высокий')
ax4.plot(reliability_universe, reliability['очень_высокий'].mf, 'b', linewidth=1.5, label='Очень высокий')
ax4.set_title('Функции принадлежности: Рейтинг надёжности')
ax4.set_xlabel('Рейтинг, баллы')
ax4.set_ylabel('Степень принадлежности')
ax4.legend()
ax4.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 1.2 База нечётких правил

База правил системы содержит 13 продукционных правил, формализующих эмпирические зависимости между характеристиками заёмщика и его кредитоспособностью. Правила сформулированы на основе экспертных знаний в области кредитного скоринга. Принципы следующие:
- Приоритет кредитной истории: Кредитная история – наиболее важный фактор. Даже при высоком доходе плохая история сильно снижает рейтинг.
- Компенсация факторов: Хорошая кредитная история может компенсировать не очень высокий доход, и наоборот, высокий доход не может полностью компенсировать плохую историю.
- Возрастные риски: Молодой возраст считается более рискованным из-за отсутствия длительной кредитной истории. Пожилой возраст также может считаться более рискованным из-за возможного снижения доходов.
- Комбинации параметров: Охват наиболее типичных комбинаций параметров, а также экстремальных случаев.

**Правило 1**: ЕСЛИ (кредитная история является отличной) И (доход является высоким) ТО (рейтинг надёжности является очень высоким).

**Правило 2**: ЕСЛИ (кредитная история является отличной) И (доход является средним) И (возраст является не молодым) ТО (рейтинг надёжности является высоким).

**Правило 3**: ЕСЛИ (кредитная история является отличной) И (доход является средним) И (возраст является молодым) ТО (рейтинг надёжности является средним).

**Правило 4**: ЕСЛИ (кредитная история является отличной) И (доход является низким) И (возраст является средним) ТО (рейтинг надёжности является средним).

**Правило 5**: ЕСЛИ (кредитная история является средней) И (доход является высоким) И (возраст является не пожилым) ТО (рейтинг надёжности является высоким).

**Правило 6**: ЕСЛИ (кредитная история является средней) И (доход является высоким) И (возраст является пожилым) ТО (рейтинг надёжности является средним).

**Правило 7**: ЕСЛИ (кредитная история является средней) И (доход является средним) ТО (рейтинг надёжности является средним).

**Правило 8**: ЕСЛИ (кредитная история является средней) И (доход является низким) ТО (рейтинг надёжности является низким).

**Правило 9**: ЕСЛИ (кредитная история является плохой) И (доход является низким ИЛИ возраст является молодым) ТО (рейтинг надёжности является очень низким).

**Правило 10**: ЕСЛИ (кредитная история является плохой) И (доход является высоким) И (возраст является средним) ТО (рейтинг надёжности является средним).

**Правило 11**: ЕСЛИ (возраст является пожилым) И (доход является низким) ТО (рейтинг надёжности является низким).

**Правило 12**: ЕСЛИ (возраст является молодым) И (доход является низким) И (кредитная история является средней) ТО (рейтинг надёжности является низким).

**Правило 13**: ЕСЛИ (кредитная история является плохой) И (доход является средним) ТО (рейтинг надёжности является низким).

Логическая связка «И» в условиях правил реализована через операцию взятия минимума (min-норма), что соответствует интерпретации конъюнкции в нечёткой логике Заде. Связка «ИЛИ» реализована через операцию взятия максимума (max-конорма).


```python
rule1 = ctrl.Rule(credit_history['отличная'] & income['высокий'], 
                  reliability['очень_высокий'],
                  label='Отличная_история_+_Высокий_доход')

rule2 = ctrl.Rule(credit_history['отличная'] & income['средний'] & ~age['молодой'], 
                  reliability['высокий'],
                  label='Отличная_история_+_Средний_доход_+_Не_молодой')

rule3 = ctrl.Rule(credit_history['отличная'] & income['средний'] & age['молодой'], 
                  reliability['средний'],
                  label='Отличная_история_+_Средний_доход_+_Молодой')

rule4 = ctrl.Rule(credit_history['отличная'] & income['низкий'] & age['средний'], 
                  reliability['средний'],
                  label='Отличная_история_+_Низкий_доход_+_Средний_возраст')

rule5 = ctrl.Rule(credit_history['средняя'] & income['высокий'] & ~age['пожилой'], 
                  reliability['высокий'],
                  label='Средняя_история_+_Высокий_доход_+_Не_пожилой')

rule6 = ctrl.Rule(credit_history['средняя'] & income['высокий'] & age['пожилой'], 
                  reliability['средний'],
                  label='Средняя_история_+_Высокий_доход_+_Пожилой')

rule7 = ctrl.Rule(credit_history['средняя'] & income['средний'], 
                  reliability['средний'],
                  label='Средняя_история_+_Средний_доход')

rule8 = ctrl.Rule(credit_history['средняя'] & income['низкий'], 
                  reliability['низкий'],
                  label='Средняя_история_+_Низкий_доход')

rule9 = ctrl.Rule(credit_history['плохая'] & (income['низкий'] | age['молодой']), 
                  reliability['очень_низкий'],
                  label='Плохая_история_+_Низкий_доход_или_Молодой')

rule10 = ctrl.Rule(credit_history['плохая'] & income['высокий'] & age['средний'], 
                   reliability['низкий'],
                   label='Плохая_история_+_Высокий_доход_+_Средний_возраст')

rule11 = ctrl.Rule(age['пожилой'] & income['низкий'], 
                   reliability['низкий'],
                   label='Пожилой_+_Низкий_доход')

rule12 = ctrl.Rule(age['молодой'] & income['низкий'] & credit_history['средняя'], 
                   reliability['низкий'],
                   label='Молодой_+_Низкий_доход_+_Средняя_история')

rule13 = ctrl.Rule(credit_history['плохая'] & income['средний'], 
                   reliability['низкий'],
                   label='Молодой_+_Низкий_доход_+_Средняя_история')
```

Создание системы управления (схема Мамдани)

```python
reliability_ctrl = ctrl.ControlSystem([
    rule1, rule2, rule3, rule4, rule5,
    rule6, rule7, rule8, rule9, rule10,
    rule11, rule12, rule13
])
```

Создание симулятора системы

```python
scoring_system = ctrl.ControlSystemSimulation(reliability_ctrl)
```

### 1.3 Выбранные методы для этапов нечеткого вывода

**Фаззификация**: Применён метод одноточечной фаззификации, при котором для каждого чёткого входного значения вычисляется степень его принадлежности ко всем определённым лингвистическим термам соответствующей переменной. Это наиболее распространённый метод, не требующий дополнительных предположений о распределении входных данных.

**Агрегация подусловий**: Использована min-норма (операция минимума) для объединения степеней истинности отдельных подусловий, соединённых связкой «И». Для связки «ИЛИ» применена max-конорма (операция максимума). Данный выбор соответствует принципу консервативной оценки: степень истинности составного условия определяется наименее истинным компонентом при конъюнкции и наиболее истинным при дизъюнкции.

**Активация заключений**: Выбран метод минимума (Mamdani), при котором выходная функция принадлежности, соответствующая заключению правила, «обрезается» на уровне вычисленной степени истинности всего условия. Альтернативный метод произведения (Larsen) не применялся, так как метод минимума лучше соответствует семантике правил в задачах качественной оценки.

**Аккумуляция**: Использована операция максимума для объединения активированных выходных нечётких множеств от всех правил. Данная операция обеспечивает учёт вклада каждого правила, при этом итоговое нечёткое множество отражает наиболее оптимистичную из всех рекомендаций (в интерпретации данной задачи – наивысшую возможную оценку надёжности, подтверждённую каким-либо правилом).

**Дефаззификация**: Применён метод центра тяжести (centroid). Выходное чёткое значение вычисляется как абсцисса центра тяжести площади под кривой результирующей функции принадлежности:

y = (∑_i μ(y_i)*y_i) / (∑_i μ(y_i)),

где суммирование проводится по всем дискретным точкам универсального множества выходной переменной. Данный метод обеспечивает плавное изменение выходного рейтинга при варьировании входных параметров и учитывает форму всей результирующей функции принадлежности, что повышает адекватность системы по сравнению с методами, учитывающими только точки максимума.

Выбор именно этой совокупности методов образует схему вывода Мамдани, которая является стандартом де-факто для систем нечёткого вывода в задачах управления и оценки, благодаря своей прозрачности, интерпретируемости и устойчивости результатов.

### 2.2 Визуализация поверхности принятия решений

Для анализа многомерных зависимостей между входными параметрами и выходным рейтингом, а также для визуализации логики работы системы нечёткого вывода были построены поверхности принятия решений. Поверхность принятия решений – это трёхмерная визуализация, отображающая зависимость выходной переменной (рейтинга надёжности) от двух входных переменных при фиксированном значении третьей переменной. Данный метод позволяет наглядно представить решающие правила системы и оценить характер взаимосвязей между параметрами.

Построение поверхности осуществлялось путём дискретизации диапазонов двух входных переменных с последующим вычислением рейтинга надёжности для всех возможных комбинаций их значений. Третья входная переменная фиксировалась на определённом уровне, характерном для типичного заёмщика. В частности, была построена поверхность для зависимости рейтинга от дохода и кредитной истории при фиксированном возрасте 45 лет, что соответствует наиболее активной и финансово устойчивой возрастной группе.

Дискретизация производилась с шагом 10 тыс. рублей для дохода (диапазон 0–200 тыс. рублей) и с шагом 5 баллов для кредитной истории (диапазон 0–100 баллов). Для каждой из 21x21 = 441 комбинации значений система нечёткого вывода вычисляла рейтинг надёжности, формируя матрицу результатов размером 21x21. Полученные данные визуализировались в виде трёхмерной поверхности, где по осям X и Y откладывались значения кредитной истории и дохода соответственно, а по оси Z – соответствующий рейтинг надёжности.

Построенная поверхность демонстрирует нелинейный характер зависимости рейтинга от входных параметров.

```python
x_income = np.arange(0, 201, 10)
y_history = np.arange(0, 101, 5)

X, Y = np.meshgrid(y_history, x_income)
Z = np.zeros_like(X, dtype=float)

fixed_age = 45

for i in range(len(x_income)):
    for j in range(len(y_history)):
        scoring_system.input['доход'] = x_income[i]
        scoring_system.input['кредитная_история'] = y_history[j]
        scoring_system.input['возраст'] = fixed_age
        
        try:
            scoring_system.compute()
            Z[i, j] = scoring_system.output['рейтинг_надёжности']
        except:
            Z[i, j] = 0

fig6 = plt.figure(figsize=(14, 10))
ax6 = fig6.add_subplot(111, projection='3d')

surf = ax6.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', alpha=0.8, linewidth=0.5)

ax6.set_xlabel('Кредитная история, баллы', fontsize=12, labelpad=15)
ax6.set_ylabel('Доход, тыс. руб.', fontsize=12, labelpad=15)
ax6.set_zlabel('Рейтинг надёжности, баллы', fontsize=12, labelpad=15)
ax6.set_title(f'Поверхность принятия решений (Возраст = {fixed_age} лет)', 
              fontsize=14, pad=20)

fig6.colorbar(surf, ax=ax6, shrink=0.5, aspect=10, label='Рейтинг надёжности')

ax6.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()

print(f"Поверхность построена для фиксированного возраста: {fixed_age} лет")
```

Наблюдаются следующие закономерности:
- Монотонное возрастание рейтинга с увеличением как дохода, так и кредитной истории, что соответствует интуитивным представлениям о кредитоспособности.
- Различная чувствительность рейтинга к входным параметрам: в области низких значений кредитной истории рейтинг остаётся низким даже при высоких доходах, что подчёркивает приоритетное значение кредитной истории в оценке надёжности заёмщика. Напротив, в области высоких значений кредитной истории рост дохода приводит к увеличению рейтинга.
- Наличие плато в области высоких значений обоих параметров (доход > 150 тыс. рублей, кредитная история > 80 баллов), где рейтинг стабилизируется, приближаясь к максимальным значениям.
- Плавность поверхности свидетельствует об отсутствии резких скачков в работе системы, что является следствием плавных переходов между лингвистическими термами и корректной работы механизма дефаззификации.
- Область максимального риска расположена вблизи начала координат (низкий доход и плохая кредитная история), где рейтинг принимает минимальные значения (0–10 баллов).

Полученная поверхность подтверждает адекватность построенной системы нечёткого вывода: система демонстрирует логически последовательное поведение, при котором улучшение характеристик заёмщика приводит к повышению рейтинга, а наиболее значимым фактором является кредитная история. Визуализация также позволяет идентифицировать области параметрического пространства, в которых система проявляет наибольшую чувствительность к изменениям входных данных, что может быть использовано для дальнейшей тонкой настройки функций принадлежности и правил вывода.


### 2.3 Анализ адекватности и устойчивости системы

Для комплексной оценки качества разработанной системы нечёткого вывода был проведён анализ адекватности и устойчивости. Анализ выполнялся на выделенном экземпляре системы с целью объективной оценки её рабочих характеристик без влияния предыдущих вычислений.

Адекватность системы оценивалась путём тестирования на пяти характерных тестовых случаях, специально подобранных для проверки корректности работы системы в различных ситуациях, включая граничные и экстремальные значения входных параметров.

```python
print("Анализ адекватности и устойчивости системы\n")

analysis_system = ctrl.ControlSystemSimulation(reliability_ctrl)

print("Базовый случай, чтобы убедиться, что новая система работает на стандартных данных")
analysis_system.input['доход'] = 100
analysis_system.input['кредитная_история'] = 70
analysis_system.input['возраст'] = 40
analysis_system.compute()
print(f"Стандартный тест: {analysis_system.output['рейтинг_надёжности']:.2f}\n")

boundary_tests = [
    {
        "name": "Очень низкие значения",
        "inputs": {"доход": 0, "кредитная_история": 0, "возраст": 18},
        "expected": "очень низкий",
        "reason": "Очень низкий доход и плохая кредитная история"
    },
    {
        "name": "Очень высокие значения", 
        "inputs": {"доход": 200, "кредитная_история": 100, "возраст": 90},
        "expected": "очень высокий",
        "reason": "Очень высокий доход и отличная кредитная история"
    },
    {
        "name": "Высокий доход, плохая история",
        "inputs": {"доход": 175, "кредитная_история": 0, "возраст": 40},
        "expected": "низкий",
        "reason": "Кредитная история важнее дохода"
    },
    {
        "name": "Низкий доход, отличная история",
        "inputs": {"доход": 0, "кредитная_история": 100, "возраст": 50},
        "expected": "средний",
        "reason": "Хорошая история компенсирует низкий доход"
    },
    {
        "name": "Средние значения",
        "inputs": {"доход": 100, "кредитная_история": 50, "возраст": 45},
        "expected": "средний",
        "reason": "Средние значения по всем параметрам"
    }
]

def get_rating_category(score):
    if score >= 80:
        return "очень высокий"
    elif score >= 60:
        return "высокий"
    elif score >= 40:
        return "средний"
    elif score >= 20:
        return "низкий"
    else:
        return "очень низкий"

boundary_results = []
print("\nРезультаты тестирования на характерных случаях:")
print("-" * 120)
print(f"{'Тест':<35} {'Доход':<8} {'История':<10} {'Возраст':<8} {'Рейтинг':<10} {'Категория':<15} {'Ожидание':<15} {'Соответствие':<10}")
print("-" * 120)

successful_tests = 0
for test in boundary_tests:
    analysis_system.input['доход'] = test['inputs']['доход']
    analysis_system.input['кредитная_история'] = test['inputs']['кредитная_история']
    analysis_system.input['возраст'] = test['inputs']['возраст']
    
    try:
        analysis_system.compute()
        score = analysis_system.output['рейтинг_надёжности']
        category = get_rating_category(score)
        
        if "высокий" in test['expected'] and "высокий" in category:
            match = "+"
        elif "низкий" in test['expected'] and "низкий" in category:
            match = "+"
        elif "средний" in test['expected'] and "средний" in category:
            match = "+"
        else:
            match = "-"
        
        successful_tests += 1
    except Exception as e:
        score = 0
        category = "ошибка вычисления"
        match = "-"
        print(f"Ошибка для теста '{test['name']}': {str(e)[:50]}")
    
    boundary_results.append({
        **test,
        "actual_score": score,
        "actual_category": category,
        "match": match
    })
    
    print(f"{test['name']:<35} "
          f"{test['inputs']['доход']:<8} "
          f"{test['inputs']['кредитная_история']:<10} "
          f"{test['inputs']['возраст']:<8} "
          f"{score:<10.2f} "
          f"{category:<15} "
          f"{test['expected']:<20} "
          f"{match:<10}")

print("-" * 120)

if successful_tests > 0:
    adequate_tests = sum(1 for r in boundary_results if r['match'] == '+')
    adequacy_percentage = (adequate_tests / successful_tests) * 100
    
    print("\nРезультат анализа адекватности:")
    print(f"Успешно выполнено тестов: {successful_tests} из {len(boundary_tests)}")
    print(f"Пройдено тестов (соответствие): {adequate_tests} из {successful_tests}")
    print(f"Процент адекватности: {adequacy_percentage:.1f}%")
else:
    print(f"\nВнимание: ни один тест не был выполнен успешно!")
```

Ключевые выводы по адекватности:
- Система успешно выполнила все 5 тестовых случаев, демонстрируя 100% успешность вычислений
- Для всех случаев полученные категории рейтинга полностью соответствуют ожидаемым
- Процент адекватности системы составил 100%, что свидетельствует о высокой степени соответствия поведения системы экспертной логике
- Система корректно обрабатывает как экстремальные (0 и 200 тыс. руб., 0 и 100 баллов), так и промежуточные значения входных параметров
- Наблюдается логическая последовательность в результатах: улучшение характеристик заёмщика приводит к повышению рейтинга надёжности

Устойчивость системы оценивалась через анализ её чувствительности к малым изменениям входных параметров. Исследование проводилось на базовом тестовом случае, представляющем стандартного заёмщика: доход 100 тыс. рублей, кредитная история 70 баллов, возраст 40 лет.

Методология анализа устойчивости:
1.	Определение базового случая с вычисленным рейтингом 70.00 баллов (категория "высокий")
2.	Последовательное варьирование каждого входного параметра при фиксированных остальных
3.	Для каждого параметра применялось изменение на ±10% от базового значения
4.	Расчёт абсолютного изменения рейтинга (Δ) для каждой вариации
5.	Определение средней чувствительности по каждому параметру
6.	Классификация уровня чувствительности на основе величины изменения рейтинга


```python
print("\nАнализ устойчивости системы")

base_case = {"доход": 100, "кредитная_история": 70, "возраст": 40}

print("\nБазовый случай (гарантированно рабочий):")
print(f"Доход: {base_case['доход']} тыс. руб.")
print(f"Кредитная история: {base_case['кредитная_история']} баллов")
print(f"Возраст: {base_case['возраст']} лет")

try:
    for key, value in base_case.items():
        analysis_system.input[key] = value
    analysis_system.compute()
    base_score = analysis_system.output['рейтинг_надёжности']
    base_category = get_rating_category(base_score)
    
    print(f"Базовый рейтинг: {base_score:.2f} баллов")
    print(f"Категория: {base_category}")
    
    print(f"\nАнализ чувствительности (±10% от базовых значений):")
    
    sensitivity_params = [
        {
            "name": "Доход",
            "key": "доход",
            "base": base_case['доход'],
            "unit": "тыс. руб.",
            "step": 10,
            "variations": [-10, 10]
        },
        {
            "name": "Кредитная история",
            "key": "кредитная_история",
            "base": base_case['кредитная_история'],
            "unit": "баллов",
            "step": 7,
            "variations": [-7, 7]
        },
        {
            "name": "Возраст",
            "key": "возраст",
            "base": base_case['возраст'],
            "unit": "лет",
            "step": 4,
            "variations": [-4, 4]
        }
    ]
    
    sensitivity_results = []
    
    for param in sensitivity_params:
        variations_scores = []
        variations_deltas = []
        
        for variation in param['variations']:
            for key, value in base_case.items():
                analysis_system.input[key] = value
            
            analysis_system.input[param['key']] = param['base'] + variation
            
            try:
                analysis_system.compute()
                score = analysis_system.output['рейтинг_надёжности']
                delta = score - base_score
                variations_scores.append(score)
                variations_deltas.append(delta)
            except Exception as e:
                variations_scores.append(base_score)
                variations_deltas.append(0)
        
        avg_abs_change = sum(abs(d) for d in variations_deltas) / len(variations_deltas)
        
        if avg_abs_change < 1.0:
            sensitivity_level = "низкая"
        elif avg_abs_change < 3.0:
            sensitivity_level = "умеренная"
        elif avg_abs_change < 5.0:
            sensitivity_level = "высокая"
        else:
            sensitivity_level = "очень высокая"
        
        sensitivity_results.append({
            "parameter": param['name'],
            "avg_abs_change": avg_abs_change,
            "sensitivity": sensitivity_level,
            "base_value": param['base'],
            "unit": param['unit']
        })
        
        print(f"\n{param['name']}:")
        print(f"Базовое значение: {param['base']} {param['unit']}")
        print(f"Изменение -{param['step']} {param['unit']}: "
              f"{variations_scores[0]:.2f} (Δ={variations_deltas[0]:+.2f})")
        print(f"Изменение +{param['step']} {param['unit']}: "
              f"{variations_scores[1]:.2f} (Δ={variations_deltas[1]:+.2f})")
        print(f"Среднее абсолютное изменение: {avg_abs_change:.2f} баллов")
        print(f"Уровень чувствительности: {sensitivity_level}")
    
    print(f"\nСводка по чувствительности системы:")
    most_sensitive = max(sensitivity_results, key=lambda x: x['avg_abs_change'])
    least_sensitive = min(sensitivity_results, key=lambda x: x['avg_abs_change'])
    
    print(f"Наиболее чувствительный параметр: {most_sensitive['parameter']} "
          f"({most_sensitive['avg_abs_change']:.2f} баллов)")
    print(f"Наименее чувствительный параметр: {least_sensitive['parameter']} "
          f"({least_sensitive['avg_abs_change']:.2f} баллов)")
    
except Exception as e:
    print(f"Ошибка при анализе чувствительности: {e}")
```

Кредитная история является наиболее значимым параметром с очень высокой чувствительностью (5.34 балла среднего изменения). Ухудшение кредитной истории на 10% приводит к существенному снижению рейтинга на 10.69 баллов, что подчёркивает критическую важность этого параметра в оценке кредитоспособности.

Доход и возраст демонстрируют низкую чувствительность в области выбранного базового случая. Изменения этих параметров на ±10% не приводят к изменению рейтинга, что указывает на наличие "плато" в функции отклика системы для данных значений.

Асимметричность реакции: система демонстрирует различную чувствительность к улучшению и ухудшению параметров. В частности, улучшение кредитной истории выше определённого порога не даёт дополнительного выигрыша в рейтинге, в то время как её ухудшение существенно снижает оценку.

Стабильность системы: для стандартных значений параметров система демонстрирует высокую устойчивость к малым изменениям дохода и возраста, что является положительной характеристикой, исключающей неоправданно резкие изменения рейтинга при незначительных колебаниях входных данных.
